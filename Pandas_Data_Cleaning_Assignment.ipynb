## Q1. Find Missing Values

Input:
df = pd.DataFrame({
    "Name":["Aman","Riya",None],
    "Age":[21,np.nan,25]
})

Output:
    Name    Age
0  False  False
1  False   True
2   True  False

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({"Name" : ['Aman','Riya',None] , "Age" : [21,np.nan,25]})
print(df.isna())
```

## Q2. Count Total Missing Values

Input:
df = pd.DataFrame({
    "Name":["Aman",None,"Raj"],
    "Age":[20,None,25]
})

Output:
Name    1
Age     1

```python
df = pd.DataFrame({
    "Name":["Aman",None,"Raj"],
    "Age":[20,None,25]
})
print(df.isnull().sum())
```

## Q3. Fill Missing Age with 0

Input:
df = pd.DataFrame({

    "Age":[20,None]
})

Output:
   Name   Age
0  Aman  20.0
1   Raj   0.0

```python
df['Age'] = df['Age'].fillna(0)
print(df)
```

## Q4. Fill Missing City with 'Unknown'

Input:
df = pd.DataFrame({
    "Name":["Aman","Raj"],
    "City":["Bhopal",None]
})

Output:
   Name     City
0  Aman   Bhopal
1   Raj  Unknown

```python
df = pd.DataFrame({
    "Name":["Aman","Raj"],
    "City":["Bhopal",None]
})
df['City'] = df['City'].fillna("Unknown")
print(df)
```

## Q6. Convert Age Column into Float

Input:
df = pd.DataFrame({"Age":[20,25,30]})

Output:
20.0, 25.0, 30.0

```python
df = pd.DataFrame({"Age":[20,25,30]})
df['Age'] = df['Age'].astype(float)
print(df)
```

## Q7. Replace Delhi with New Delhi

Input:
df = pd.DataFrame({

})

```python
df = pd.DataFrame({
    "City":["Delhi","Mumbai","Delhi"]
})
df['City'] = df['City'].replace("Delhi","New Delhi")
print(df)
```

## Q13. Mini Data Cleaning Task

Input:
data = {
    "Name":["Aman","Raj","Aman",None],
    "Age":[20,None,20,25],
    "City":["Bhopal","Delhi","Bhopal","Indore"]
}

Task:
1. Fill missing Age with mean.
2. Remove duplicate rows.
3. Display final DataFrame.

```python
data = {
    "Name":["Aman","Raj","Aman",None],
    "Age":[20,None,20,25],
    "City":["Bhopal","Delhi","Bhopal","Indore"]
}
df = pd.DataFrame(data)
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df.drop_duplicates())
```
